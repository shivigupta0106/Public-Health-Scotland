```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from datetime import datetime
from random import randint
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib notebook
%matplotlib inline
%matplotlib qt
```

```python
#Creating figure and subplot objects
# Figure is the graph, and subplot is the axis of the graph
fig = plt.figure()
ax = fig.add_subplot()
```

```python
#Using random generated data to create animated graph

#Create lists to hold data
time_series = []
value_a_series = []
value_b_series = []

#Creating function to get dummy data 

def generate_dummy_data():
    time = datetime.now()
    value_a = randint(-100, 100)
    value_b = randint(-10, 10)

    time_series.append(time)
    value_a_series.append(value_a)
    value_b_series.append(value_b)

#Fuction to create an animated graph

def animate(i):

    generate_dummy_data()

#As the axis of the graph can constantly change, they need to
# be cleared and redrawn each time new data has been added.

ax.clear()

plt.title('An animated graph')
plt.xlabel('Time')
plt.ylabel('values')

#plot contents of the lists of data

ax.plot(time_series, value_a_series, label='value_a_series')
ax.plot(time_series, value_b_series, label='value_b_series')

#add legend
plt.legend()

fig.autofmt_xdate()

#Animating the graph
#call the animate function every 1000 milliseconds, which is every second.
#It needs to be passed the fig object as the first parameter, and then the
#name of the function to be called.
ani = animation.FuncAnimation(fig,
                              animate,
                              interval=1000)
                            

#plot display
plt.show()
```

```python
# Creating figure and subplot objects
fig, ax = plt.subplots()

# Lists to hold data
time_series = []
value_a_series = []
value_b_series = []

# Function to generate dummy data
def generate_dummy_data():
    time = datetime.now()
    value_a = randint(-100, 100)
    value_b = randint(-10, 10)
    time_series.append(time)
    value_a_series.append(value_a)
    value_b_series.append(value_b)

# Function to create an animated graph
def animate(i):
    generate_dummy_data()
    ax.clear()
    ax.plot(time_series, value_a_series, label='Value A Series')
    ax.plot(time_series, value_b_series, label='Value B Series')
    ax.set_title('An animated graph')
    ax.set_xlabel('Time')
    ax.set_ylabel('Values')
    ax.legend()
    fig.autofmt_xdate()

# Animating the graph
ani = animation.FuncAnimation(fig, animate, interval=1000)

# Plot display
plt.show()
```
