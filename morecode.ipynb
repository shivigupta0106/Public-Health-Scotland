```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from datetime import datetime
from random import randint
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
```

```python
#%matplotlib notebook
#%matplotlib inline
#%matplotlib qt

from IPython.display import HTML
```

```python
url1 = 'https://www.opendata.nhs.scot/dataset/0d57311a-db66-4eaa-bd6d-cc622b6cbdfa/resource/a5f7ca94-c810-41b5-a7c9-25c18d43e5a4/download/weekly_ae_activity_20240616.csv'
df_week_AE = pd.read_csv(url1)
df_week_AE.head()
```

```python
# Convert WeekEndingDate to datetime format
df_week_AE['WeekEndingDate'] = pd.to_datetime(df_week_AE['WeekEndingDate'], format='%Y%m%d')
#The pd.to_datetime function converts the WeekEndingDate column from a string or integer format (YYYYMMDD) to a pandas datetime object.

# Extract Year and Month
df_week_AE['Year'] = df_week_AE['WeekEndingDate'].dt.year
df_week_AE['Month'] = df_week_AE['WeekEndingDate'].dt.month
df_week_AE['Day'] = df_week_AE['WeekEndingDate'].dt.day

# Display the updated DataFrame
df_week_AE.head()
```

```python
# Sort the DataFrame by Date
df1 = df_week_AE.sort_values(by='WeekEndingDate')
df1.head()

df2= df1[['WeekEndingDate','NumberOfAttendancesEpisode']]
```

```python
df2.head()
```

```python
fig, ax= plt.subplots()
line, = ax.plot_date(df2['WeekEndingDate'], df2['NumberOfAttendancesEpisode'])


def animate(i):
    line.set_data(df2['WeekEndingDate'][:i],df2['NumberOfAttendancesEpisode'][:i])
    ax.relim()
    ax.autoscale_view()
    return line,
```

```python
ani= animation.FuncAnimation(fig,animate,frames=1000, interval=1,blit=True)
HTML(ani.to_jshtml())
```

```python
# Creating figure and subplot objects
fig, ax = plt.subplots()

# Lists to hold data
time_series = []
value_a_series = []
value_b_series = []

# Function to generate dummy data
def generate_dummy_data():
    time = datetime.now()
    value_a = randint(-100, 100)
    value_b = randint(-10, 10)
    time_series.append(time)
    value_a_series.append(value_a)
    value_b_series.append(value_b)

# Function to create an animated graph
def animate(i):
    generate_dummy_data()
    ax.clear()
    ax.plot(time_series, value_a_series, label='Value A Series')
    ax.plot(time_series, value_b_series, label='Value B Series')
    ax.set_title('An animated graph')
    ax.set_xlabel('Time')
    ax.set_ylabel('Values')
    ax.legend()
    fig.autofmt_xdate()

# Animating the graph
ani = animation.FuncAnimation(fig, animate, interval=1000)

# Plot display
plt.show()
```

```python
# Initialize the plot
fig, ax = plt.subplots()
line, = ax.plot([], [], 'b-', linewidth=2, label='Number of Attendances')
ax.set_xlim(df1['WeekEndingDate'].min(), df1['WeekEndingDate'].max())
ax.set_ylim(0, df1['NumberOfAttendancesEpisode'].max() + 50)
ax.set_xlabel('Date')
ax.set_ylabel('Number of Attendances')
ax.set_title('Time Simulation of Number of Attendances')
ax.legend()

# Data for the animation
dates = df1['WeekEndingDate'].values
attendance = df1['NumberOfAttendancesEpisode'].values

# Function to update the graph
def update(frame):
    current_dates = dates[:frame]
    current_attendance = attendance[:frame]
    line.set_data(current_dates, current_attendance)
    return line,

# Create the animation
anim = FuncAnimation(fig, update, frames=np.arange(1, len(dates) + 1), interval=200, blit=True)

# Show the plot
plt.show()
```
